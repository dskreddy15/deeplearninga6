```python
from google.colab import drive
drive.mount('/content/drive')

from tables import file
from pandas.core.indexes.multi import F
import pandas as pd

file1 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.benign.csv')
file1['label'] = 'benign'
file2 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.gafgyt.combo.csv')
file2['label'] = 'gafgyt combo'
file3 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.gafgyt.junk.csv')
file3['label'] = 'gafgyt junk'
file4 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.gafgyt.scan.csv')
file4['label'] = 'gafgyt scan'
file5 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.gafgyt.tcp.csv')
file5['label'] = 'gafgyt tcp'
file6 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.gafgyt.udp.csv')
file6['label'] = 'gafgyt udp'
file7 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.mirai.ack.csv')
file7['label'] = 'mirai ack'
file8 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.mirai.scan.csv')
file8['label'] = 'mirai scan'
file9 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.mirai.syn.csv')
file9['label'] = 'mirai syn'
file10 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.mirai.udp.csv')
file10['label'] = 'mirai udp'
file11 = pd.read_csv('/content/drive/MyDrive/archive (1)/2/2.mirai.udpplain.csv')
file11['label'] = 'mirai udpplain'

file0 = pd.concat([file1, file2])
del file1,file2
file0 = pd.concat([file0,file3])
del file3
file0 = pd.concat([file0,file4])
del file4
file0 = pd.concat([file0,file5])
del file5
file0 = pd.concat([file0,file6])
del file6
file0 = pd.concat([file0,file7])
del file7
file0 = pd.concat([file0,file8])
del file8
file0 = pd.concat([file0,file9])
del file9
file0 = pd.concat([file0,file10])
del file10
file0 = pd.concat([file0,file11])
del file11

file0['label'].value_counts()

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
file0.label = le.fit_transform(file0.label)

import numpy as np

# assume your DataFrame is named 'df'
file0 = file0.replace([np.inf, -np.inf], np.nan)  # replace infinite values with NaN
#da = da[da.abs() < np.finfo(np.float64).max]  # remove all rows containing values larger than max float

file0 = file0.dropna()

# print the new DataFrame shape to verify the infinite or too large values have been removed
print(file0.shape)

data = file0.iloc[:,:-1]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(file0.iloc[:,:-1])

scaled_features

X = pd.DataFrame(scaled_features, columns = data.columns)
y = file0.iloc[:,-1]

print(X.shape)
print(y.shape)

from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=100)
Xrus,Yrus = rus.fit_resample(X,y)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Xrus,Yrus, test_size=0.2, random_state=100)

X_train.shape

y_train.shape

# Running Machine Learning Models

#1. Naive Bayes
import time

from sklearn.naive_bayes import BernoulliNB

ml1 = BernoulliNB()

start_ttime = time.time()

ml1.fit(X_train, y_train)

end_ttime = time.time()

train_time = end_ttime - start_ttime

ml1.score(X_train,y_train)

ml1.score(X,y)

start_time = time.time()

y_pred_nb = ml1.predict(X)

end_time = time.time()

test_time = end_time - start_time

print("TT", test_time)

print("TT1", train_time)
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy Score : {:.3f}".format(accuracy_score(y, y_pred_nb)))
print("\nClassification Report :")
print(classification_report(y, y_pred_nb))

import sklearn
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

cm = confusion_matrix(y, y_pred_nb)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

from sklearn.metrics import log_loss

y_pred_proba = ml1.predict_proba(X)
logloss = log_loss(y, y_pred_proba)
print(logloss)

from sklearn.metrics import precision_score
macro_score_rf = precision_score(y,y_pred_nb,average='macro')
micro_score_rf = precision_score(y,y_pred_nb,average='micro')
weighted_score_rf = precision_score(y,y_pred_nb,average='weighted')

#precision scores
print(macro_score_rf)
print(micro_score_rf)
print(weighted_score_rf)

from sklearn.metrics import recall_score
rmacro_score_rf = recall_score(y,y_pred_nb,average='macro')
rmicro_score_rf = recall_score(y,y_pred_nb,average='micro')
rweighted_score_rf = recall_score(y,y_pred_nb,average='weighted')

#precision scores
print(rmacro_score_rf)
print(rmicro_score_rf)
print(rweighted_score_rf)

from sklearn.tree import DecisionTreeClassifier

ml2 = DecisionTreeClassifier(random_state=100)

start_ttime = time.time()

ml2.fit(X_train, y_train)

end_ttime = time.time()

train_time = end_ttime - start_ttime


print(ml2.score(X_train,y_train))
print(ml2.score(X,y))

start_time = time.time()

y_pred_dt = ml2.predict(X)


end_time = time.time()

test_time = end_time - start_time

print("TT", test_time)

print("TT1", train_time)

print("Accuracy Score : {:.3f}".format(accuracy_score(y, y_pred_dt)))
print("\nClassification Report :")
print(classification_report(y, y_pred_dt))

cm = confusion_matrix(y, y_pred_dt)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

y_pred_proba = ml2.predict_proba(X)
logloss = log_loss(y, y_pred_proba)
print(logloss)

from sklearn.metrics import precision_score
macro_score_rf = precision_score(y,y_pred_dt,average='macro')
micro_score_rf = precision_score(y,y_pred_dt,average='micro')
weighted_score_rf = precision_score(y,y_pred_dt,average='weighted')

#precision scores
print(macro_score_rf)
print(micro_score_rf)
print(weighted_score_rf)

from sklearn.metrics import recall_score
rmacro_score_rf = recall_score(y,y_pred_dt,average='macro')
rmicro_score_rf = recall_score(y,y_pred_dt,average='micro')
rweighted_score_rf = recall_score(y,y_pred_dt,average='weighted')

#precision scores
print(rmacro_score_rf)
print(rmicro_score_rf)
print(rweighted_score_rf)

from sklearn.ensemble import RandomForestClassifier

ml6 = RandomForestClassifier(random_state=100)

start_ttime = time.time()

ml6.fit(X_train, y_train)

end_ttime = time.time()

train_time = end_ttime - start_ttime

print(ml6.score(X_train,y_train))
print(ml6.score(X,y))

start_time = time.time()

y_pred = ml6.predict(X)


end_time = time.time()

test_time = end_time - start_time

print("TT", test_time)

print("TT1", train_time)

from sklearn.metrics import classification_report, accuracy_score

print("Accuracy Score : {:.3f}".format(accuracy_score(y, y_pred)))
print("\nClassification Report :")
print(classification_report(y, y_pred))

import sklearn
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

from sklearn.metrics import log_loss

y_pred_proba = ml6.predict_proba(X)
logloss = log_loss(y, y_pred_proba)
print(logloss)

from sklearn.metrics import precision_score
macro_score_rf = precision_score(y,y_pred,average='macro')
micro_score_rf = precision_score(y,y_pred,average='micro')
weighted_score_rf = precision_score(y,y_pred,average='weighted')

#precision scores
print(macro_score_rf)
print(micro_score_rf)
print(weighted_score_rf)

from sklearn.metrics import recall_score
rmacro_score_rf = recall_score(y,y_pred,average='macro')
rmicro_score_rf = recall_score(y,y_pred,average='micro')
rweighted_score_rf = recall_score(y,y_pred,average='weighted')

#precision scores
print(rmacro_score_rf)
print(rmicro_score_rf)
print(rweighted_score_rf)

from sklearn.tree import ExtraTreeClassifier

ml3 = ExtraTreeClassifier(random_state=100)

start_ttime = time.time()

ml3.fit(X_train, y_train)

end_ttime = time.time()

train_time = end_ttime - start_ttime


print(ml3.score(X_train,y_train))
print(ml3.score(X,y))

start_time = time.time()

y_pred_et = ml3.predict(X)

end_time = time.time()

test_time = end_time - start_time

print("TT", test_time)
print("TT1", train_time)
print("Accuracy Score : {:.3f}".format(accuracy_score(y, y_pred_et)))
print("\nClassification Report :")
print(classification_report(y, y_pred_et))

cm = confusion_matrix(y, y_pred_et)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

y_pred_proba = ml3.predict_proba(X)
logloss = log_loss(y, y_pred_proba)
print(logloss)

from sklearn.metrics import precision_score
macro_score_rf = precision_score(y,y_pred_et,average='macro')
micro_score_rf = precision_score(y,y_pred_et,average='micro')
weighted_score_rf = precision_score(y,y_pred_et,average='weighted')

#precision scores
print(macro_score_rf)
print(micro_score_rf)
print(weighted_score_rf)

from sklearn.metrics import recall_score
rmacro_score_rf = recall_score(y,y_pred_et,average='macro')
rmicro_score_rf = recall_score(y,y_pred_et,average='micro')
rweighted_score_rf = recall_score(y,y_pred_et,average='weighted')

#precision scores
print(rmacro_score_rf)
print(rmicro_score_rf)
print(rweighted_score_rf)
```
