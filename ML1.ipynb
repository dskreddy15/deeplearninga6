```python
def correlation(dataset, threshold):
  col_corr = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i, j]) > threshold:
        colname = corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=100)
Xrus,Yrus = rus.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Xrus,Yrus, test_size=0.2, random_state=100)

X_train.shape

y_train.shape
```

```python
corr_features = correlation(X_train,0.9)
len(set(corr_features))
```

```python
corr_features
```

```python
X_train = X_train.drop(corr_features,axis = 1)
X_test = X_test.drop(corr_features, axis = 1)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

k = 10

selector = SelectKBest(f_classif, k=k) 
selector.fit(X_train, y_train)

selected_indices = selector.get_support(indices=True)
selected_features = X_train.columns[selected_indices]

scores = selector.scores_
plt.bar(range(len(scores)), scores)
plt.xlabel('Feature Index')
plt.ylabel('F-Score')
plt.show()

top_scores = sorted(scores, reverse=True)[:k]
top_indices = [list(scores).index(score) for score in top_scores]
top_features = X_train.columns[top_indices]
plt.bar(range(k), top_scores)
plt.xticks(range(k), top_features, rotation=90)
plt.xlabel('Feature')
plt.ylabel('F-Score')
plt.show()
```

```python
best_table = pd.DataFrame({'Feature': top_features , 'Importance': top_scores })
print(best_table)
```

```python
# Running Machine Learning Models

#1. Naive Bayes
import time

from sklearn.naive_bayes import BernoulliNB

ml1 = BernoulliNB()

start_ttime = time.time()

ml1.fit(X_train, y_train)

end_ttime = time.time()

train_time = end_ttime - start_ttime

ml1.score(X_train,y_train)

ml1.score(X_test,y_test)

start_time = time.time()

y_pred_nb = ml1.predict(X_test)

end_time = time.time()

test_time = end_time - start_time

print("TT", test_time)

print("TT1", train_time)
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy Score : {:.3f}".format(accuracy_score(y_test, y_pred_nb)))
print("\nClassification Report :")
print(classification_report(y_test, y_pred_nb))

import sklearn
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred_nb)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

from sklearn.metrics import log_loss

y_pred_proba = ml1.predict_proba(X_test)
logloss = log_loss(y_test, y_pred_proba)
print(logloss)
```

```python
from itertools import cycle
from sklearn.metrics import roc_curve, auc
y_pred_proba = ml1.predict_proba(X_test)
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(len(label_counts)):
    fpr[i], tpr[i], _ = roc_curve(y_test == ml1.classes_[i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
colors = cycle(['blue', 'red', 'green','aqua', 'darkorange', 'orange','fuchsia', 'lime','magenta'])
for i, color in zip(range(len(label_counts)), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0,1.2])
plt.ylim([0.0,1.2])
plt.ylabel('Recall')
plt.xlabel('Fall-out (1-Specificity)')
plt.title('Receiver Operating Characteristic (ROC) for Naive Bayes model')
plt.legend(loc="lower right")

plt.show()
```

```python
from sklearn.metrics import precision_score
macro_score_rf = precision_score(y_test,y_pred_nb,average='macro')
micro_score_rf = precision_score(y_test,y_pred_nb,average='micro')
weighted_score_rf = precision_score(y_test,y_pred_nb,average='weighted')

#precision scores
print(macro_score_rf)
print(micro_score_rf)
print(weighted_score_rf)

from sklearn.metrics import recall_score
rmacro_score_rf = recall_score(y_test,y_pred_nb,average='macro')
rmicro_score_rf = recall_score(y_test,y_pred_nb,average='micro')
rweighted_score_rf = recall_score(y_test,y_pred_nb,average='weighted')

#precision scores
print(rmacro_score_rf)
print(rmicro_score_rf)
print(rweighted_score_rf)
```
