```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
file0.label = le.fit_transform(file0.label)

file0
```

```python
import numpy as np

# assume your DataFrame is named 'df'
file0 = file0.replace([np.inf, -np.inf], np.nan)  # replace infinite values with NaN
#da = da[da.abs() < np.finfo(np.float64).max]  # remove all rows containing values larger than max float

file0 = file0.dropna()

# print the new DataFrame shape to verify the infinite or too large values have been removed
print(file0.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(file0.iloc[:,:-1])
```

```python
X = scaled_features
y = file0.iloc[:,-1]

print(X.shape)
print(y.shape)
```

```python
y.value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=100)
Xrus,Yrus = rus.fit_resample(X,y)
```

```python
Xrus.shape
```

```python
from sklearn.decomposition import PCA

# assume your DataFrame is named 'df'
# assume your target variable is named 'target'


# instantiate the PCA object with n_components = number of features
pca = PCA(n_components=Xrus.shape[1])

# fit and transform the data
X_pca = pca.fit_transform(Xrus,Yrus)

# print the explained variance ratio for each principal component
print(pca.explained_variance_ratio_)

# plot the explained variance ratio for each principal component
import matplotlib.pyplot as plt
plt.plot(range(1,Xrus.shape[1]+1), pca.explained_variance_ratio_)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()

```
