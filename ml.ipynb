```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
```

```python
X[numerical_columns] = Scaler.fit_transform(X[numerical_columns])
```

```python
from imblearn.over_sampling import SMOTE
rus = SMOTE(sampling_strategy = 'minority',random_state = 100)
Xrus,Yrus = rus.fit_resample(X,y)

Yrus.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Xrus, Yrus, test_size=0.2, random_state=100)
```

```python
import matplotlib.pyplot as plt

corr_matrix = X_train.corr()
plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.show()
```

```python
corr_matrix
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

# evaluate the performance of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#best features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

selector = SelectKBest(f_classif, k=10)

selector.fit_transform(X_train, y_train)

X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)


selector.scores_
```

```python
#GridSearchCV

lr_param_list = {'penalty': ['l1', 'l2', 'none'],
    'C': [0.1, 1, 10],
    'solver': ['newton-cg', 'liblinear'],
    'max_iter': [100, 500, 1000]}

gnb_param_list = {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]}

rf_param_list = { 'n_estimators': [5,10],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]}

kn_param_list = {'n_neighbors': [ 5, 10],
    'weights': ['uniform', 'distance']}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search_lr = GridSearchCV(clf, lr_param_list, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit the GridSearchCV object to the data
grid_search_lr.fit(X_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters: ", grid_search_lr.best_params_)

# Print the best accuracy score
print("Best mean_squared_error score: ", grid_search_lr.best_score_)

# Make predictions on the test set using the best model
best_model = grid_search_lr.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns

sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt = '.2f')

print(classification_report(y_test,y_pred))
```
